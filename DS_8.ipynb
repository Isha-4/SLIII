```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=UserWarning, module="seaborn.axisgrid")
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(int(df['age'].mean())).astype('int')
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])

df['deck'] = df['deck'].astype('object')  
df['deck'] = df['deck'].fillna('Unknown')
df.isnull().sum()
```

```python
numeric_columns=['survived',	'pclass',	'age',	'sibsp',	'parch',	'fare']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numeric_columns])
plt.xticks(rotation=45)
plt.title('Boxplot for Numeric Columns')
plt.show()
```

```python
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_mask = (df[numeric_columns] < lower_bound) | (df[numeric_columns] > upper_bound)

df_no_outliers = df[~outlier_mask.any(axis=1)]

print("Dataset before outlier removal:", df.shape)
print("Dataset after outlier removal:", df_no_outliers.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_no_outliers[numeric_columns])
plt.xticks(rotation=45)
plt.title('Boxplot for Numeric Columns')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

numeric_columns=['survived','pclass','age','sibsp','parch','fare']
scaler = MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df[numeric_columns].head() 
```
